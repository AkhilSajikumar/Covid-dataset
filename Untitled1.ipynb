```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
```

```python
datas=pd.read_csv('covid_worldwide.csv')
```

```python
datas
```

```python
Covid = datas.iloc[0:80, 0:8]
```

```python
Covid
```

```python
Covid.dropna(inplace=True)
```

```python
Covid['Total Cases'] = Covid['Total Cases'].replace({',': ''}, regex=True).astype(float)
Covid['Total Deaths'] = Covid['Total Deaths'].replace({',': ''}, regex=True).astype(float)
```

```python
data_subset = Covid[['Total Cases', 'Total Deaths']]
```

```python
scaler = StandardScaler()
data_scaled = scaler.fit_transform(data_subset)
```

```python
inertia = []
for k in range(1, 11):
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(data_scaled)
    inertia.append(kmeans.inertia_)

# Plotting the Elbow Method for Optimal k
plt.figure(figsize=(8, 6))
plt.plot(range(1, 11), inertia, marker='o', color='b')
plt.title('Elbow Method for Optimal k')
plt.xlabel('Number of Clusters (k)')
plt.ylabel('Inertia')
plt.grid(True)
plt.show()
```

```python
kmeans = KMeans(n_clusters=3, random_state=42)
kmeans.fit(data_scaled)

# Step 6: Add the cluster labels to the data
Covid['Cluster'] = kmeans.labels_

# Step 7: Visualize the clusters
plt.figure(figsize=(8, 6))
plt.scatter(data_scaled[:, 0], data_scaled[:, 1], c=Covid['Cluster'], cmap='viridis', marker='o')
plt.title('KMeans Clustering of Total Cases and Total Deaths')
plt.xlabel('Scaled Total Cases')
plt.ylabel('Scaled Total Deaths')
plt.colorbar(label='Cluster')
plt.grid(True)
plt.show()

# Optionally, display the cluster centers
cluster_centers = scaler.inverse_transform(kmeans.cluster_centers_)
print("Cluster Centers (original scale):")
print(cluster_centers)

# Display the first few rows of the data with cluster labels
print(Covid.head())
```
